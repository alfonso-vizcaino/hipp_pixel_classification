# SVM and Randrom Forest Models


```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.svm import LinearSVC, SVC, SVR
from sklearn.model_selection import RandomizedSearchCV, cross_val_predict, cross_val_score

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import SGDClassifier
from scipy.stats import expon, reciprocal

import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, roc_auc_score
from sklearn.metrics import precision_score, recall_score, f1_score

from joblib import dump, load
```

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..', 'util'))
if module_path not in sys.path:
    sys.path.append(module_path)

import numpy as np
import pandas as pd

from datasetGenerator import plot_precision_recall_vs_threshold
from datasetGenerator import plot_roc_curve


%load_ext autoreload
%autoreload 2

np.random.seed(37)
```

## Load dataset

```python
trainDataset = pd.read_csv('../data/trainDatasetValues_R35.csv', index_col=0)
trainLabels = pd.read_csv('../data/trainDatasetLabels_R35.csv', index_col=0).values.ravel()

testDataset = pd.read_csv('../data/testDatasetValues_R35.csv', index_col=0).values
testLabels = pd.read_csv('../data/testDatasetLabels_R35.csv', index_col=0).values.ravel()

print("trainDataset.shape\n", trainDataset.shape)
print("")
print("trainLabels.shape\n", trainLabels.shape)
print("")
print("testDataset.shape\n", testDataset[:, 3:].shape)
print("")
print("testLabels.shape\n", testLabels.shape)
print("")
```

Remove debug columns from dataset

```python
X_train = trainDataset.values[:, 3:]
y_train = trainLabels


X_test = testDataset[:, 3:]
y_test = testLabels
```

## Create a StandarScaler to normalize data

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Random Forest Classifier

Create classifier, train it and get its performance

```python
forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(X_train, y_train)
y_hat = forest_clf.predict(X_test)
y_probas_forest = cross_val_predict(forest_clf, X_test, y_test, cv=15, method="predict_proba")
```

Get roc_curve metrics

```python
y_scores_forest = y_probas_forest[:, 1] # score = proba of positive class
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_test, y_scores_forest)
```

Plot roc ruve

```python
tfc_au = roc_auc_score(y_test, y_scores_forest)

plot_roc_curve(fpr_forest, tpr_forest, "Random Forest {:.4f}".format(tfc_au))
plt.legend(loc="lower right")
plt.show()
```

Compute Random Forest's confusion Matrix, Precision Score, Recall Score and F1 Score

```python

cm = confusion_matrix(y_test, y_hat)
print(cm)
print("")

ps = precision_score(y_test, y_hat)
print("Precision Score {:.4f}".format(ps))

rs = recall_score(y_test, y_hat)
print("Recall Score {:.4f}".format(rs))

f1 = f1_score(y_test, y_hat)
print("F1 Score {:.4f}".format(f1))
```

## SVM Lineal

Create classifier, train it, get its metrics and plot the result

```python
lsvm_clf = LinearSVC(penalty = "l2",loss="squared_hinge", C=182, max_iter=8900, random_state=42)
#LinearSVC(C=1.9, loss="hinge", max_iter=8900, random_state=42 )

print(lsvm_clf)
lsvm_clf.fit(X_train, y_train)

y_hat = lsvm_clf.predict(X_test)
y_scores_lsvm = cross_val_predict(lsvm_clf, X_test, y_test, cv=15, method="decision_function")
lsvm_au = roc_auc_score(y_test, y_scores_lsvm)

fpr_lsvm, tpr_lsvm, thresholds = roc_curve(y_test, y_scores_lsvm)

plot_roc_curve(fpr_forest, tpr_forest, "Random Forest {:.4f}".format(tfc_au))
plot_roc_curve(fpr_lsvm, tpr_lsvm, "SVM Lineal {:.4f}".format(lsvm_au))
plt.legend(loc="lower right")
plt.show()

cm = confusion_matrix(y_test, y_hat)
print(cm)
print("")

ps = precision_score(y_test, y_hat)
print("Precision Score {:.4f}".format(ps))

rs = recall_score(y_test, y_hat)
print("Recall Score {:.4f}".format(rs))

f1 = f1_score(y_test, y_hat)
print("F1 Score {:.4f}".format(f1))
```

## SVM with polynomial features

Create classifier, train it, get its metrics and plot the result

```python
polynomial_svm_clf = Pipeline((
    ("poly_features", PolynomialFeatures(degree=2)),
    ("svm_clf", LinearSVC(C=172, loss="squared_hinge", penalty="l2", dual=False, random_state=42))
))

#polynomial_svm_clf = Pipeline((
#    ("poly_features", PolynomialFeatures(degree=2)),
#    ("svm_clf", LinearSVC(C=23.5, loss="squared_hinge", penalty="l2", dual=False, random_state=42))
#))

print(polynomial_svm_clf)
polynomial_svm_clf.fit(X_train, y_train)

y_hat = polynomial_svm_clf.predict(X_test)
y_scores_psvm = cross_val_predict(polynomial_svm_clf, X_test, y_test, cv=15, method="decision_function")

psvm_au = roc_auc_score(y_test, y_scores_psvm)
fpr_psvm, tpr_psvm, thresholds = roc_curve(y_test, y_scores_psvm)

plot_roc_curve(fpr_forest, tpr_forest, "Random Forest {:.4f}".format(tfc_au))
plot_roc_curve(fpr_lsvm, tpr_lsvm, "SVM lineal {:.4f}".format(lsvm_au))
plot_roc_curve(fpr_psvm, tpr_psvm, "SVM poly char {:.4f}".format(psvm_au))
plt.legend(loc="lower right")
plt.show()


cm = confusion_matrix(y_test, y_hat)
print(cm)
print("")

ps = precision_score(y_test, y_hat)
print("Precision Score {:.4f}".format(ps))

rs = recall_score(y_test, y_hat)
print("Recall Score {:.4f}".format(rs))

f1 = f1_score(y_test, y_hat)
print("F1 Score {:.4f}".format(f1))
```

## SVM with polynomial kernel

Create classifier, train it, get its metrics and plot the result

```python
polK_svm_clf = SVC(kernel="poly", degree=3, coef0=40, C=2.1, random_state=42)

print(polK_svm_clf)
polK_svm_clf.fit(X_train, y_train)

y_hat = polK_svm_clf.predict(X_test)
y_scores_pksvm = cross_val_predict(polK_svm_clf, X_test, y_test, cv=15, method="decision_function")

pksvm_au = roc_auc_score(y_test, y_scores_pksvm)
fpr_pksvm, tpr_pksvm, thresholds = roc_curve(y_test, y_scores_pksvm)

plot_roc_curve(fpr_forest, tpr_forest, "Random Forest {:.4f}".format(tfc_au))
plot_roc_curve(fpr_lsvm, tpr_lsvm, "SVM lineal {:.4f}".format(lsvm_au))
plot_roc_curve(fpr_psvm, tpr_psvm, "SVM poly char {:.4f}".format(psvm_au))
plot_roc_curve(fpr_pksvm, tpr_pksvm, "SVM poly kernel {:.4f}".format(pksvm_au))
plt.legend(loc="lower right")
plt.show()

cm = confusion_matrix(y_test, y_hat)
print(cm)
print("")

ps = precision_score(y_test, y_hat)
print("Precision Score {:.4f}".format(ps))

rs = recall_score(y_test, y_hat)
print("Recall Score {:.4f}".format(rs))

f1 = f1_score(y_test, y_hat)
print("F1 Score {:.4f}".format(f1))
```

## SVM with RBF kernel

Create classifier, train it, get its metrics and plot the result

```python
grbf_svm_clf = SVC(kernel="rbf", gamma = "scale", coef0 = 1.09, C = 182, random_state=42)

print(grbf_svm_clf)
grbf_svm_clf.fit(X_train, y_train)

y_hat = grbf_svm_clf.predict(X_test)
y_scores_gsvm = cross_val_predict(grbf_svm_clf, X_test, y_test, cv=15, method="decision_function")

gsvm_au = roc_auc_score(y_test, y_scores_gsvm)
fpr_gsvm, tpr_gsvm, thresholds = roc_curve(y_test, y_scores_gsvm)

plot_roc_curve(fpr_forest, tpr_forest, "Random Forest {:.4f}".format(tfc_au))
plot_roc_curve(fpr_lsvm, tpr_lsvm, "SVM lineal {:.4f}".format(lsvm_au))
plot_roc_curve(fpr_psvm, tpr_psvm, "SVM poly char {:.4f}".format(psvm_au))
plot_roc_curve(fpr_pksvm, tpr_pksvm, "SVM poly kernel {:.4f}".format(pksvm_au))
plot_roc_curve(fpr_gsvm, tpr_gsvm, "SVM RBF kernel {:.4f}".format(gsvm_au))
plt.legend(loc="lower right")
plt.show()

cm = confusion_matrix(y_test, y_hat)
print(cm)
print("")

ps = precision_score(y_test, y_hat)
print("Precision Score {:.4f}".format(ps))

rs = recall_score(y_test, y_hat)
print("Recall Score {:.4f}".format(rs))

f1 = f1_score(y_test, y_hat)
print("F1 Score {:.4f}".format(f1))
```
