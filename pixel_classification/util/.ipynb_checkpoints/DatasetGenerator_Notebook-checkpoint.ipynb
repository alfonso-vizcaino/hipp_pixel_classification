## 2) Load raw data 

Features are contained in a csv file that contain 14 columns. Every row represents a pixel, every column a pixel feature.

Training-set, test-set and validation-set consitute 70% , 20% and 10% respectively, from the raw dataset.

```python
trainSet, testSet, valSet = getDataSet("../data/all_features_r29.csv")
```

### 2.1) Get ahold of the data and its labels separately.
### 2.2) Make sure that label date only contains 1's for pixels representing a hippocampus pixel and 0's otherwise for a binary classification


```python


#Make binary classification
trainSet[1].values[np.where(trainSet[1].values > 1)] =1

#Make binary classification
testSet[1].values[np.where(testSet[1].values > 1)] = 1

#Make binary classification
valSet[1].values[np.where(valSet[1].values > 1)] = 1
```

## 3) Verify that the amounts of the train, test and validation set match each other and that they are properly split

### 3.1) Verify that the amount of pixel features should be the same as the amount of pixel labels across the different sets

```python
print("")
print("TRAINING DATASET")
print("Train data shape, Rows={}, Columns={}".format(*trainSet[0].values.shape)) 
print("Train label shape, Rows={}, Columns={}".format(*trainSet[1].values.shape))

print("")
print("TEST DATASET")
print("Test data shape, Rows={}, Columns={}".format(*testSet[0].values.shape)) 
print("Test label shape, Rows={}, Columns={}".format(*testSet[1].values.shape))


print("")
print("VAL DATASET")
print("Val data shape, Rows={}, Columns={}".format(*valSet[0].shape)) 
print("Val label shape, Rows={}, Columns={}".format(*valSet[1].shape))
```

### 3.2) Verify that every image contained in the set should have the same amount of hippocampus and non-hippocampus pixels

#### 3.2.1) Get ahold of the colum that contain the image name
#### 3.2.2) Obtain the amount of occurences the image is present in the set
#### 3.2.3) Verify that the set contains -roughly- the same amount of hippocampus and non-hippocampus pixels accross the different sets


```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)

print("")
print("TRAINING DATASET")

col = trainSet[0].values[:,0]
uniques, c  = np.unique(col, return_counts=True)

headers=["Image Name", "Hippocampus Pixels", "Non-Hippocampus Pixels", "Total Pixel Count", "Ratio"]
table = pd.DataFrame()

for img_name, count in zip(uniques, c) :
    img_rows = trainSet[0]['Source'] == img_name
    img_x_values = trainSet[0].loc[ img_rows ] 
    img_y_values = trainSet[1].loc[ img_rows ] 
    
    tp = img_x_values.loc[img_y_values['Y'] > 0 ].shape[0]
    tn = img_x_values.loc[img_y_values['Y'] == 0 ].shape[0]
    
    table = table.append(pd.DataFrame([[img_name, tp, tn, count, tp/tn]], columns=headers), ignore_index= True)
    #print(img_name+ " has " +str(count) + " records; tp count="+ str(tp) + "; tn=" + str( tn) + "; diff=" + str(tp/tn))
    
table
```

```python

print("")
print("TEST DATASET")
col = testSet[0].values[:,0]
uniques, c  = np.unique(col, return_counts=True)

table = pd.DataFrame()

for img_name, count in zip(uniques, c) :
    img_rows = testSet[0]['Source'] == img_name
    img_x_values = testSet[0].loc[ img_rows ] 
    img_y_values = testSet[1].loc[ img_rows ] 
    
    tp = img_x_values.loc[img_y_values['Y'] > 0 ].shape[0]
    tn = img_x_values.loc[img_y_values['Y'] == 0 ].shape[0]
    
    table = table.append(pd.DataFrame([[img_name, tp, tn, count, tp/tn]], columns=headers), ignore_index= True)
    #print(img_name+ "=" +str(count) + "; tp count="+ str(tp) + "; tn=" + str( tn) + "; diff=" + str(tp/tn))
    
table
```

```python

print("")
print("VAL DATASET")

col = valSet[0].values[:,0]
uniques, c  = np.unique(col, return_counts=True)

table = pd.DataFrame()

for img_name, count in zip(uniques, c) :
    img_rows = valSet[0]['Source'] == img_name
    img_x_values = valSet[0].loc[ img_rows ] 
    img_y_values = valSet[1].loc[ img_rows ] 
    
    tp = img_x_values.loc[img_y_values['Y'] > 0 ].shape[0]
    tn = img_x_values.loc[img_y_values['Y'] == 0 ].shape[0]
    
 
    table = table.append(pd.DataFrame([[img_name, tp, tn, count, tp/tn]], columns=headers), ignore_index= True)
    #print(img_name+ "=" +str(count) + "; tp count="+ str(tp) + "; tn=" + str( tn) + "; diff=" + str(tp/tn))
    
table
```
